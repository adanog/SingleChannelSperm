Load libraries

```python
import tifffile as tiff
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the TIFF file
tiff_file = '/content/drive/MyDrive/2024/estudiantes/Rodrigo Escobedo/SingleChannelSperm/Mibe 30 + Iono 3-1_crop.tif'
with tiff.TiffFile(tiff_file) as tif:
    img_stack = tif.asarray()  # Carga el archivo como un array de NumPy

# Check the dimensions of the TIFF file
print(f"TIFF file dimensions: {img_stack.shape}")

# Assume the file has three dimensions: (t, y, x)
t, y, x = img_stack.shape
print(f"Dimensions: t = {t}, y = {y}, x = {x}")

# Example of how to access a specific frame in the time dimension (t)
frame_index = 0  # Change to any index within the range [0, t-1]
frame = img_stack[frame_index, :, :]

# Show information about the selected frame
print(f"Frame {frame_index} dimensions: {frame.shape}")
```

```python

# Calculate the average value of the image for each time point
# Assuming the image stack is of shape (t, y, x)
avg_values = np.mean(img_stack, axis=(1, 2))  # Average over y and x dimensions

# Plot the average values over time
plt.figure(figsize=(10, 6))
plt.plot(avg_values, marker='o', linestyle='-', color='b')
plt.title('Average Image Value over Time')
plt.xlabel('Time (t)')
plt.ylabel('Average Pixel Value')
plt.grid(True)
plt.show()
```
